```python
import random

patente = []
marca = []
modelo = []
precio = []
multas = []
monto_multa = []
fecha_multa = []
fecha_registro = []
dueño = []

while True:
    print('''
    ---------------------------------------------------------
    Bienvenido al sistema de la automotora
              Auto Seguro
    ---------------------------------------------------------
    ''')
    opc = input('''
    ---------------------------------------------------------
    Por favor ingrese una opción
    1) Grabar
    2) Buscar
    3) Imprimir certificados
    4) Salir
    ---------------------------------------------------------
    ''')

    while not opc.isnumeric():
        opc = input("Por favor ingresar un dígito numérico: \n")

    if int(opc) == 1:
        a = input("Ingrese el nombre del dueño: \n")
        dueño.append(str(a))

        b = input("Ingrese una patente: \n")
        while len(b) != 6:
            print("La patente debe tener 6 caracteres.")
            b = input("Ingrese una patente: \n")
        patente.append(str(b))

        c = input("Ingrese una marca: \n")
        marca.append(str(c))

        d = input("Ingrese un modelo: \n")
        modelo.append(str(d))

        e = input("Ingrese un precio: \n")
        while int(e) < 5000000:
            print("El precio debe ser mayor a 5 millones\n")
            e = input("Ingrese un precio: \n")
        precio.append(int(e))

        f = input("Ingrese las multas asociadas al vehículo: \n")
        multas.append(str(f))

        g = input("Ingrese el monto de la multa en dígitos numéricos: \n")
        monto_multa.append(int(g))

        h = input("Ingrese la fecha de la multa: \n")
        fecha_multa.append(str(h))

        i = input("Ingrese la fecha de registro: \n")
        fecha_registro.append(str(i))

    elif int(opc) == 2:
        buscar = input("Ingrese la patente que desea buscar: \n")
        if buscar in patente:
            buscar2 = patente.index(buscar)
            print(f'''
            Patente:           {patente[buscar2]}
            Marca:             {marca[buscar2]}
            Modelo:            {modelo[buscar2]}
            Precio:            {precio[buscar2]}
            Multa(s):          {multas[buscar2]}
            Monto de multa:    {monto_multa[buscar2]}
            Fecha de multa:    {fecha_multa[buscar2]}
            Fecha de registro: {fecha_registro[buscar2]}
            Dueño:             {dueño[buscar2]}
            ''')
        else:
            print("La patente no fue encontrada.")

    elif int(opc) == 3:
        buscar = input("Ingrese la patente para buscar sus certificados: \n")
        if buscar in patente:
            buscar2 = patente.index(buscar)
            print(f'''
            Certificado de emisiones
                   - APROBADO
            Anotaciones vigentes
            - Patente:          {patente[buscar2]}
            - Marca:            {marca[buscar2]}
            - Precio:           {precio[buscar2]}
            - Fecha de multa:   {fecha_multa[buscar2]}
            - Multa:            {multas[buscar2]}
            ''')
    elif int(opc) == 4:
      print("Hasta la proxima :D")
      break
```
